```python
import numpy as np
import random

def newtonRaphson(f, diff):
    # Defining initial value
    x0 = random.randint(-2e9, 2e9)

    while True:
        x1 = x0 - f(x0)/diff(x0)
        if (abs(x1 - x0) <= 1e-9):
            break
        x0 = x1
    
    return x1
```

```python
def f(x):
    return x**3

def diff(x):
    return 3*(x**2)

print("Root: {:.9f}".format(newtonRaphson(f, diff)))
```

```python
def f1(x):
    return x**3 + x**2 + 3

def diff1(x):
    return 3*x**2 + 2*x

print("Root: {:.9f}".format(newtonRaphson(f1, diff1)))
```
