```python
import numpy as np

# Defining a numpy array that stores data
arr = np.array([[1, 2, 3, 4, 5, 6, 7], [1.5 , 3.8, 6.7, 9.0, 11.2, 13.6, 16]])
n = len(arr[0])

# Sum of values of x and y
sum_x = arr[0].sum()  
sum_y = arr[1].sum()

# Sum of values of xi*yi
sum_xy = 0
for i in range(n):
    sum_xy = sum_xy + arr[0][i]*arr[1][i]

# Sum of squares of xi
sum_x2 = 0
for i in range(n):
    sum_x2 = sum_x2 + arr[0][i]**2

# Slope of linear curve
m = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)

# y intercept of linear curve
b = (sum_y - m*sum_x)/n

print("slope: ", m)
print("y-intercept: ", b)
```
