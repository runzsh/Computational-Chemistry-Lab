```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LJ(sig, eps, r):
    return 4*eps*((sig/r)**12 - (sig/r)**6)
```

```python
r_min = 3
r_max = 15

x = np.linspace(r_min, r_max, 100, True)
```

```python
y = LJ(3.4, 1, x)
```

```python
plt.plot(x, y)
plt.xlabel("r/sigma")
plt.ylabel("Vlj/eps")
```
